# Data Analytics I
1. Create a Linear Regression Model using Python/R to predict home prices using Boston Housing Dataset (https://www.kaggle.com/c/boston-housing). 
2. The Boston Housing dataset contains information about various houses in Boston through different parameters. There are 506 samples and 14 feature variables in this dataset. 
The objective is to predict the value of prices of the house using the given features.

```python
import numpy as nm
import pandas as pd
import sklearn
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston.keys()
```

```python
boston.data.shape
```

```python
# Print column names
print(boston.feature_names)
```

```python
data = pd.DataFrame(data= boston.data , columns=boston.feature_names )
data.head()
```

```python
data.describe()
```

```python
data.keys()
```

```python
data.isnull().sum()
```

```python
x = data[['CRIM', 'ZN', 'INDUS']]
y = data['AGE']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.9, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
rsme = nm.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error is: ")
print(rsme)
```

```python
print("Accuracy of Training is: ")
lr.score(x_train, y_train)
```

```python
print("Accuracy of Testing is: ")
lr.score(x_test, y_test)
```
